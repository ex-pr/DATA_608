# 1. Data Preparation

## 1.1 Load data, html table

During the data loading phase of this salary analysis, we performed a series of meticulous web scraping operations to collect current and relevant salary data (March, 2024) for a wide range of roles in the data and analytics domain. This process entailed accessing specific, curated web pages on Zippia, a well-known career and employment resource, to obtain salary information for a variety of key positions, including Data Analysts, Data Scientists, Data Engineers, Data Architects, Business Analysts, and AI Specialists. Each page featured detailed salary tables, which were then systematically loaded into separate dataframes. To ensure clarity and consistency across datasets, each dataframe now includes a 'Job_role' column that labels each entry with its corresponding job title, such as 'Data Analyst' or 'Data Scientist'.


```python
# @title
# libraries
import requests
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import requests
from bs4 import BeautifulSoup
import warnings
warnings.filterwarnings("ignore")
```

```python
# @title
# URL of the webpage to scrape 'Data Analyst' salary
url = "https://www.zippia.com/data-analyst-jobs/salary/"

# Make a request to get the HTML content of the page
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Find the first table in the HTML
table = soup.find('table')

# Read the table into a pandas DataFrame
analyst_df = pd.read_html(str(table))[0]

# Add a 'Job_role' column to the DataFrame
analyst_df['Job_role'] = 'Data Analyst'

# Display the DataFrame
print(analyst_df.head())
```

## 1.2 Summary Statistics

Our analysis includes salary datasets for a wide range of roles in the data and analytics industry, including Data Analysts, Data Scientists, Data Engineers, Data Architects, Business Analysts, and AI Specialists. Each dataset was sourced from up-to-date web pages (Zippia.com), ensuring that the salary information is relevant and accurate (March, 2024).

The datasets for all roles are uniformly structured, with 150 observations and 5 key variables each, with no missing values. This uniformity in data structure simplifies comparative analysis across positions, allowing for a better understanding of salary trends, geographic variations, and the impact of expertise levels on compensation in the data and analytics industry.



## 1.3 Data cleaning

The process began with the consolidation of datasets obtained through targeted web scraping, with the goal of creating a single view of salary trends across these roles.

To facilitate comprehensive analysis, we first combined individual datasets into a single dataframe. This step mirrored the merging process used with economic datasets, resulting in a unified dataset that encompasses the wide range of roles in the data and analytics industry.

The 'Avg. Salary', 'Hourly Rate', and 'Job Count' columns were converted from string representations with commas and dollar signs to integer data types. This standardization was critical for performing quantitative analyses on salary data across the board. Next, we removed non-numeric characters from the salary and job count columns to ensure that the data was clean and ready for analysis. This step was similar to removing irrelevant or redundant columns from economic datasets in order to streamline the data for analysis.


The cleaned and consolidated dataset included data from **306 observations** across the various roles, which were now standardized with **6 variables**, including cleaned salary information.
The average salary across all roles was around 84,741 dollars, demonstrating the lucrative nature of the data and analytics industry.
The compensation for the roles varied, with AI Specialists earning an average of 131,538, and Data Analysts earning 65,541 dollars, reflecting the industry's demand and specialization.

```python
# @title
# Concatenate the dataframes
final_df = pd.concat([ai_df, analyst_df, architect_df, business_df, engineer_df, scientist_df], ignore_index=True)

# Remove dollar signs and commas, then convert to integer
final_df['Avg. Salary'] = final_df['Avg. Salary'].replace('[\$,]', '', regex=True).astype(float)
final_df['Hourly Rate'] = final_df['Hourly Rate'].replace('[\$,]', '', regex=True).astype(float)
final_df['Job Count'] = final_df['Job Count'].replace('[\,]', '', regex=True).astype(float)

# Display the first few rows of the final dataframe
print(final_df)
```

# 2. Data analysis

The first visualization, a boxplot, depicts the salary distribution for data practitioners in various job roles based on March, 2024 data, revealing significant differences in earning potential. **AI Specialists**, in particular, are expected to command higher median salaries of approximately **132,780** dollars due to the high demand for their advanced machine learning and artificial intelligence skills. This is consistent with industry trends emphasizing the ability to leverage data for predictive analytics and strategic insights. Data Architects are the next highest earners, with a significant median salary, demonstrating the industry's recognition of the importance of designing robust data systems that are scalable, secure, and performance optimized. In contrast, the **Data Analyst** role, which is frequently used as an entry point into the data field, has a broader salary range with the median salary of **64,892 dollars**. This suggests a dynamic market in which data analysis skills are required but compensation varies depending on experience, industry, and task complexity. Other roles, such as Data Engineers and Business Analysts, have salary ranges that reflect the importance of their jobs in facilitating data accessibility and translating data insights into actionable business strategies, respectively. These roles are critical links in the data value chain, connecting raw data management to strategic decision-making.

The second visualization uses a geographical lens to highlight the top and bottom five states by average salary across all job roles. States known for their technological and business hubs, such as **California and New York** (with averages exceeding **100,000** dollars for roles such as AI Specialists and Data Engineers), unsurprisingly dominate the rankings, owing to the concentration of tech firms and startups willing to invest heavily in top-tier data talent. Washington and New Jersey are among the highest-paying states. Washington's high average salaries are due to its strong tech sector presence, which includes both major corporations and startups, whereas New Jersey benefits from its proximity to New York City as well as its concentration of pharmaceutical and finance industries, both of which place a high value on data analytics for business growth and innovation. The variation in salaries across states, such as Maryland, reflects a balance between the presence of federal agencies and contractors near Washington, D.C., which require data expertise, and the local cost of living.

On the other end of the spectrum, states like **Arkansas and North Dakota** have significantly lower average salaries (with figures such as **60,000** dollars for Data Analysts), which could be due to a smaller tech presence or different industrial focuses. This disparity is not only due to the value placed on data roles, but it also interacts with factors such as cost of living, state-level economic strategies, and educational resources. **Hawaii**, despite its low ranking in average salary, presents a unique case. Factors like the high cost of living, isolated geographical location, and an economy historically reliant on tourism rather than tech innovation, could explain why salaries for data roles don’t match those of the mainland tech hubs. However, as remote work grows and tech initiatives take root, these figures could evolve.

The analysis of this data not only highlights the financial value of data practitioners, but it also provides a narrative about the United States' economic, educational, and industrial landscape. The clear distinction between the earning potential of various roles and locations provides useful information for those navigating the field. It suggests that professionals, particularly those in entry-level positions or living in low-wage states, should consider upskilling, specialization, and possibly relocation to increase their earning potential. Furthermore, it encourages industry leaders and policymakers to address these disparities, possibly by promoting tech education, supporting the growth of local industries, and creating incentives to attract and retain data talent in underrepresented areas.

```python
# @title
# Calculate the median salary for each job role
median_salaries = final_df.groupby('Job_role')['Avg. Salary'].median().sort_values()

# The job roles with the lowest and highest median salaries
lowest_salary_role = median_salaries.idxmin()
highest_salary_role = median_salaries.idxmax()

# Soft green for the highest salaries and a softer shade of red for the lowest salaries
highest_salary_color = '#7FC6FF'  # pale blue
lowest_salary_color = '#F96267'  # light red

# Color palette with a neutral color for all roles except for the lowest and highest
palette = {role: 'grey' if role not in [lowest_salary_role, highest_salary_role] else lowest_salary_color if role == lowest_salary_role else highest_salary_color for role in median_salaries.index}

# Reordering the roles based on the average salary to have the roles with the highest salaries at the top
roles_by_salary = median_salaries.sort_values(ascending=False).index

# Box plot with the specified order and enhanced palette
plt.figure(figsize=(12, 8))
ax = sns.boxplot(
    x='Avg. Salary',
    y='Job_role',
    data=final_df,
    order=roles_by_salary,
    palette=palette  # Custom palette
)

# Annotations for the highest and lowest average salaries
for role in [lowest_salary_role, highest_salary_role]:
    role_median = median_salaries[role]
    ax.text(role_median, roles_by_salary.get_loc(role), f'${role_median:,.0f}',
            color='black', ha='center', va='center',
            bbox=dict(facecolor='white', edgecolor='none', alpha=0.5))

plt.title('Distribution of State Average Salaries by Job Role', fontsize=16)
plt.xlabel('Average Salary ($)', fontsize=14)
plt.ylabel('Job Role', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='x')

plt.tight_layout()
plt.show()
```

```python
# @title
# Calculate the average salary for each state
state_salary = final_df.groupby('State')['Avg. Salary'].mean().sort_values()

# Select the top 5 and bottom 5 states
top_states = state_salary.tail(5).index.tolist()
bottom_states = state_salary.head(5).index.tolist()

# Define colors for each job role
ai_color = '#7FC6FF'
data_architect_color = '#C9D0EA'
data_engineer_color = '#DAC4D5'
data_scientist_color = '#E7B6C1'
business_analyst_color = '#F2A5AB'
analyst_color = '#F96267'

# Creating the custom_palette
custom_palette = {
    'AI Specialist': ai_color,
    'Data Analyst': analyst_color,
    'Data Scientist': data_scientist_color,
    'Data Architect': data_architect_color,
    'Business Analyst': business_analyst_color,
    'Data Engineer': data_engineer_color
}

# Combine top and bottom states
selected_states = top_states + bottom_states

# Filter the dataset for the selected states and all roles
selected_df = final_df[final_df['State'].isin(selected_states)]

# Define order of job roles based on the average salary for each, from highest to lowest
role_order = ['AI Specialist', 'Data Architect', 'Data Engineer', 'Data Scientist', 'Business Analyst', 'Data Analyst']
# Sort the roles based on the desired order
selected_df['Job_role'] = pd.Categorical(selected_df['Job_role'], categories=role_order, ordered=True)

# Sort states based on average salary from the combined list of top and bottom states
# Create a dictionary to map states to their respective mean salaries
state_order = {state: salary for state, salary in zip(state_salary.index, state_salary.values)}
# Sort the selected states using the dictionary, putting top states first
selected_states_sorted = sorted(selected_states, key=lambda x: state_order[x], reverse=True)

# Sort the dataframe by the state order
selected_df['State'] = pd.Categorical(selected_df['State'], categories=selected_states_sorted, ordered=True)

# State-Specific Salary Variations for Job Roles
plt.figure(figsize=(14, 6))
bar_plot = sns.barplot(x='State', y='Avg. Salary', hue='Job_role',data=selected_df.sort_values('State'), palette=custom_palette, errorbar=None)
plt.title('Top and Bottom 5 States by Average Salary across All Job Roles', fontsize=16)
plt.xlabel('State', fontsize=14)
plt.ylabel('Average Salary ($)', fontsize=14)

# Adjust the legend location and title
plt.legend(title='Job Role', loc='upper right', bbox_to_anchor=(1.15, 1))

# Separate the top and bottom states visually by adding a vertical line
plt.axvline(4.5, color='grey', linewidth=1, linestyle='dotted')

# Find the maximum Y value to place the text below the title
max_salary = selected_df['Avg. Salary'].max()
buffer = max_salary * 0.005  # Create a small buffer to lower the text

plt.text(4.5 -1.2, max_salary - buffer, '← Highest Average Salaries',
         fontsize=12, color='dimgrey', ha='center')
plt.text(4.5 + 1.2, max_salary - buffer, 'Lowest Average Salaries →',
         fontsize=12, color='dimgrey', ha='center')

plt.tight_layout()

plt.show()
```
